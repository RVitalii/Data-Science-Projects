## Model

The chosen model falls within the category of linear models, akin to linear regression, and is widely employed for solving classification problems. Spam detection is a classical illustration of such problems, where the goal is to categorize emails as either spam or non-spam. In this context, a simple binomial logistic regression is well-suited.

The fitted model assigns probabilities of an email being spam to each input vector of feature values. These output probabilities can then be used to identify spam with a certain level of confidence.

Logistic regression is composed of two fundamental elements:

1. **The Simple Logistic Function:** 
$$p(x) = \frac{e^x}{1+e^x}.$$

2. **Linear Regression (in matrix form):**
$$Y = X\beta + \epsilon.$$

Combining the two components results in the logistic regression equation:
$$p(X) = \frac{e^{X\beta + \epsilon}}{1+e^{X\beta + \epsilon}}.$$

This outcome can be interpreted as a log of odds. The model can be expressed as:
$$log\left(\frac{p(X)}{1 - p(X)}\right) = X\beta + \epsilon.$$

It's important to note that the left-hand side of this equation represents the logit function, which is the inverse of the logistic function. It serves as the quantile function of the logistic distribution. Consequently, the model's output is a conditional probability of the binary dependent variable given a random vector of independent variables, which can be either discrete or continuous.

## Binomial Logistic Regression Assumptions

1. Independent errors.
2. Errors are homoscedastic.
3. Dependent variables are binary.
4. There is no multicollinearity between independent variables.
5. The dependet variables are exogenous, meaning they are not influenced by the dependent variables.
5. It is possible to establish linear relationship between independent variables and log odds.
6. There is no outliers in data.

## Results

```python
# imports
from io import StringIO
from pandas import read_csv
from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RepeatedKFold
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
from statsmodels.api import Logit
```

```python
# constants
UCIREPO_ID = 94
KFOLD_SID = 1
KFOLD_SPLITS = 10
KFOLD_REPEATS = 1
SCORING_METHOD = 'accuracy'
SELECTION_METHOD = 'forward'
SELECTION_TOL = 0.0001
FEATURES_TO_SELECT = 'auto'
```

```python
# fetch dataset 
spambase = fetch_ucirepo(id=UCIREPO_ID)
```

```python
# extract variables and scale independent variables
X = spambase.data.features
y = spambase.data.targets.Class

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# specify model
model = LogisticRegression()
cv = RepeatedKFold(n_splits=KFOLD_SPLITS, n_repeats=KFOLD_REPEATS, random_state=KFOLD_SID)
selector = SequentialFeatureSelector(
    estimator=model, cv=cv, scoring=SCORING_METHOD, direction=SELECTION_METHOD, 
    n_features_to_select=FEATURES_TO_SELECT, tol=SELECTION_TOL
    )

# fit selection
fitted_selector = selector.fit(X_scaled, y)

# extract selected features
selected_features = spambase.data.features.columns[fitted_selector.get_support()]
```

```python
# getting data for model
X_selected = spambase.data.features[selected_features.drop(['word_freq_3d', 'word_freq_650', 'word_freq_telnet'])]
y_selected = spambase.data.targets
```

```python
# fitting logistic regression with stats models 
sm_model = Logit(y_selected, X_selected)
sm_model_fitted = sm_model.fit()
```

```python
#printing models summary
print(sm_model_fitted.summary())
```

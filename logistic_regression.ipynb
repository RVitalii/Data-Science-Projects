# Spam Prediction

## Introduction
This project applies Principal Component Analysis (PCA) and Logistic Regression to classify spam emails from the UCI Spambase dataset. Unlike traditional approaches that rely on full-dimensional data, PCA helps reduce dimensionality while preserving relevant information, leading to improved model efficiency and performance. By automating feature selection through PCA and cross-validation, we ensure a streamlined, data-driven approach to classification.

With a focus on feature selection, model fitting, and assumption checks, the project aims to develop a robust spam prediction model. The dataset comprises 4601 instances with 57 continuous and 1 nominal class label attributes, sourced from the UCI Machine Learning Repository.

Through automation and advanced modeling, the project seeks to enhance the accuracy of spam identification, contributing to the effective management of email communication and information security challenges.

## Goal
The objective is to develop and train a binomial logistic regression model for the purpose of identifying spam.

## Project Structure

This project consist of the following parts:

1. **Feature Engineering and Preprocessing**
* The dataset is retrieved from the UCI Machine Learning Repository.
* Data is split into training and test sets to evaluate model performance.
* Features are standardized using StandardScaler to improve PCA and model accuracy.
* There are no outliers due to the inherent characteristics of the input data.

2. **Dimensionality Reduction with PCA**
* Principal Component Analysis (PCA) is applied to the data to reduce the feature space while retaining as much variance as possible.
* A custom function determines the optimal number of PCA components using cross-validation, ensuring the best trade-off between dimensionality reduction and predictive power.

3. **Logistic Regression Model Training and Evaluation**
* A Logistic Regression classifier is trained on the transformed PCA data.
* The model's performance is evaluated using a confusion matrix, accuracy scores, precision, recall, and F1 scores to provide a comprehensive assessment of its classification effectiveness.

4. **ROC Curve Analysis and Optimal Threshold Selection**
* The ROC curve and AUC score are computed to analyze model discrimination ability.
* An optimal decision threshold is selected based on maximizing the difference between True Positive Rate (TPR) and False Positive Rate (FPR), rather than relying on the default 0.5 threshold.
* Predictions are recalculated using this optimized threshold to improve classification performance.

5. **Visualization and Insights**
* The ROC curve is plotted to visualize model performance.
* A Threshold vs. TPR/FPR graph helps understand the effect of different decision boundaries.
* Key takeaways and conclusions are drawn, reflecting on the impact of PCA and optimal thresholding in improving classification accuracy.

## Model

The chosen model falls within the category of linear models, akin to linear regression, and is widely employed for solving classification problems. Spam detection is a classical illustration of such problems, where the goal is to categorize emails as either spam or non-spam. In this context, a simple binomial logistic regression is well-suited.

The fitted model assigns probabilities of an email being spam to each input vector of feature values. These output probabilities can then be used to identify spam with a certain level of confidence.

Logistic regression is composed of two fundamental elements:

1. **The Simple Logistic Function:** 
$$p(x) = \frac{e^x}{1+e^x}.$$

2. **Linear Regression (in matrix form):**
$$Y = X\beta + \epsilon.$$

Combining the two components results in the logistic regression equation:
$$p(X) = \frac{e^{X\beta + \epsilon}}{1+e^{X\beta + \epsilon}}.$$

This outcome can be interpreted as a log of odds. The model can be expressed as:
$$log\left(\frac{p(X)}{1 - p(X)}\right) = X\beta + \epsilon.$$

It's important to note that the left-hand side of this equation represents the logit function, which is the inverse of the logistic function. It serves as the quantile function of the logistic distribution and transforms the probability, which ranges between 0 and 1, into a continuous range from negative infinity to positive infinity. As a result, the model's output is a conditional probability of the binary dependent variable (e.g., spam or not) given a random vector of independent variables, which may be either discrete or continuous.

To convert the log-odds back to probabilities, we apply the logistic function (the inverse of the logit), which outputs values between 0 and 1.

## Binomial Logistic Regression Assumptions

1. Independent errors.
2. Errors are homoscedastic.
3. Dependent variables are binary.
4. There is no multicollinearity between independent variables.
5. The independet variables are exogenous, meaning they are not influenced by the dependent variable.
6. It is possible to establish a linear relationship between independent variables and log odds
7. There are no outliers in the data.

## Key Model Performance Measures

#### **1. Precision** (Positive Predictive Value)  
The proportion of correctly predicted positive instances out of all instances predicted as positive:  

$$
\text{Precision} = \frac{TP}{TP + FP}
$$  

Where:  
- **TP (True Positives)** are correctly predicted positive cases.  
- **FP (False Positives)** are incorrectly predicted positive cases.  

High precision means fewer false positives. It is useful when **false positives are costly** (e.g., spam detection).  

#### **2. Recall** (Sensitivity or True Positive Rate)  
The proportion of actual positives that were correctly identified:  

$$
\text{Recall} = \frac{TP}{TP + FN}
$$  

Where:  
- **TP (True Positives)** are correctly predicted positive cases.  
- **FN (False Negatives)** are missed positive cases (incorrectly classified as negative).  

High recall means fewer false negatives. It is important when **missing positive cases is critical** (e.g., disease detection).  

#### **3. F1-Score** (Balance of Precision & Recall)  
The harmonic mean of **Precision** and **Recall**, providing a single metric that balances both:  

$$
F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$  

Where:  
- **Precision** is the proportion of correctly predicted positives.  
- **Recall** is the proportion of actual positives correctly identified.  

The F1-score is useful when there is **an imbalance** between positive and negative cases. A higher F1-score indicates a better trade-off between precision and recall.

#### **4. False Positive Rate (FPR)**  
The proportion of actual negative cases that were incorrectly predicted as positive:  

$$
\text{FPR} = \frac{FP}{FP + TN}
$$  

Where:  
- **FP (False Positives)** are incorrectly predicted positive cases.  
- **TN (True Negatives)** are correctly predicted negative cases.  

Low FPR is crucial when you want to avoid **false alarms**. It is especially important in situations where **false positives could be costly** (e.g., recommending a non-spam email as spam).

## Results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc
)
```

```python
UCIREPO_ID = 94
TEST_SPLIT_SIZE = 0.25
RANDOM_STATE = 42
SPAM_CLASS = 1
NON_SPAM_CLASS = 0
```

```python
spambase = fetch_ucirepo(id=UCIREPO_ID)
```

```python
def get_pca_column_names(n):
    col_names = [f'PCA_{i}' for i in range(1, n + 1)]

    return col_names

def get_optimal_pca_components(x_train_data, y_train_data, model, cv=5, scoring='precision', debug=False):
    best_score = 0
    best_components = 0
    features_number = x_train_data.shape[1] + 1

    for n in range(1, features_number):
        pca = PCA(n_components=n)
        x_train_data_pca = pca.fit_transform(x_train_data)
        score = np.mean(cross_val_score(
            model, x_train_data_pca, y_train_data, cv=cv, scoring=scoring
            )
        )

        if debug:
            explained_variance = np.sum(pca.explained_variance_ratio_)
            print(f"Number of components={n}, CV Score={score:.4f}, Explained Variance={explained_variance:.4f}")

        if score > best_score:
            best_score = score
            best_components = n

    return best_components
```

```python
X, y = spambase.data.features, spambase.data.targets
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=TEST_SPLIT_SIZE, random_state=RANDOM_STATE)

y_train = y_train.squeeze()
y_test = y_test.squeeze()

scaler = StandardScaler()
X_train_standardized = scaler.fit_transform(X_train)
X_test_standardized = scaler.transform(X_test)

X_train_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)
X_test_standardized = pd.DataFrame(X_test_standardized, columns=X_test.columns)

logistic_regression = LogisticRegression()
optimal_pca_number = get_optimal_pca_components(X_train_standardized, y_train, logistic_regression, debug=False)

pca = PCA(n_components=optimal_pca_number)
X_train_standardized_pca = pca.fit_transform(X_train_standardized)
X_test_standardized_pca = pca.transform(X_test_standardized)

X_train_standardized_pca = pd.DataFrame(X_train_standardized_pca, columns=get_pca_column_names(optimal_pca_number))
X_test_standardized_pca = pd.DataFrame(X_test_standardized_pca, columns=get_pca_column_names(optimal_pca_number))

logistic_regression_fitted = logistic_regression.fit(X_train_standardized_pca, y_train)

logistic_regression_train_score = logistic_regression_fitted.score(X_train_standardized_pca, y_train)
logistic_regression_test_score = logistic_regression_fitted.score(X_test_standardized_pca, y_test)

y_prob = logistic_regression.predict_proba(X_test_standardized_pca)[:, SPAM_CLASS]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]

optimal_prediction_test = (y_prob >= optimal_threshold).astype(int)
```

```python
explained_variance = np.sum(pca.explained_variance_ratio_)

print("PCA Summary:")
print(f"Number of components {optimal_pca_number}\nExplained Variance   {explained_variance:.2f}")
```

```python
print(f'score (train): {logistic_regression_train_score:.2f}')
print(f'score (test):  {logistic_regression_test_score:.2f}')
```

```python
print(F'Optimal Threshold: {optimal_threshold:.2f}')
```

```python
cm_logreg = confusion_matrix(
    y_test, logistic_regression_fitted.predict(X_test_standardized_pca)
    )
cm_optimal = confusion_matrix(
    y_test, optimal_prediction_test
    )

cm_logreg_df = pd.DataFrame(
    cm_logreg, columns=['Predicted Negative', 'Predicted Positive'],
    index=['Actual Negative', 'Actual Positive']
    )
cm_optimal_df = pd.DataFrame(
    cm_optimal, columns=['Predicted Negative', 'Predicted Positive'],
    index=['Actual Negative', 'Actual Positive']
    )

print("Confusion Matrix for Default Threshold:")
print(cm_logreg_df)
print("\nConfusion Matrix for Optimal Threshold:")
print(cm_optimal_df)
```

```python
report_logreg = classification_report(
    y_test, logistic_regression_fitted.predict(X_test_standardized_pca)
    )
report_optimal = classification_report(
    y_test, optimal_prediction_test
    )

print("Classification Report for Default Threshold:")
print(report_logreg)
print("\nClassification Report for Optimal Threshold:")
print(report_optimal)
```

## Discussion
* The optimal number of PCA components, that maximise the precision, is 50 out of a total of 57 features, capturing 97% of the explained variance in the data. This means the last seven components contribute little to improving the model's performance. However, the difference in precision scores between the one-component model and the selected model is only around 9%. Therefore, reducing the number of components would result in only a minimal loss of performance.
* The training score, defined as the proportion of correct predictions out of the total predictions, is comparable to the test score, which is approximately 92%.
* However, the confusion matrix reveals that even with high accuracy scores, the model does not produce a fully reliable spam filter. With the optimized threshold, 44 non-spam emails were misclassified as spam, while 39 spam emails were incorrectly labeled as non-spam. In contrast, the default threshold is more conservative, resulting in fewer false positives (32) but a higher number of false negatives (62). While the default threshold prioritizes minimizing false positives, the optimized threshold reduces the total number of misclassifications, achieving a better balance between false positives and false negatives.
* Although the recall for spam predictions (Class 1) improves in the optimal threshold model (due to better identification of spam cases), the precision for Class 1 (spam) decreases slightly. However, since Class 0 (Non-Spam) is much more frequent in the data, it is crucial to maximize precision for Class 0 to avoid misclassifying spam as non-spam. As a result, the precision for non-spam (Class 0) improves, and the weighted average F1-score increases by 1% in the optimal threshold model.
* The ROC curve visualizes the trade-off between False Positive Rate (FPR) and True Positive Rate (TPR) for different classification thresholds. The optimal threshold is typically selected by maximizing Youden's J statistic (TPR - FPR). In the best-case scenario, an AUC of 1 is achieved when the ROC curve follows the leftmost and topmost edges of the plot, meaning FPR = 0 and TPR = 1. The right plot of FPR and TPR against thresholds indicates that a threshold of 0.4 maximizes this difference.

## Conclusion  

1. **PCA Selection**:  
   Using 50 components out of 57 captures 97% of the variance, with only a small performance loss (~9%) compared to a one-component model.

2. **Model Performance**:  
   The model generalizes well, with training and test accuracy around **92%**.

3. **Threshold Optimization**:  
   The **optimized threshold** reduces misclassifications by balancing false positives (44 non-spam as spam) and false negatives (39 spam as non-spam). The **default threshold** minimizes false positives but increases false negatives.

4. **Precision-Recall Tradeoff**:  
   The **optimized threshold** improves recall for spam (Class 1) but slightly lowers precision for spam. Maximizing precision for **non-spam (Class 0)** is crucial due to its higher prevalence, leading to a **1% improvement in the weighted F1-score**.

5. **ROC Curve & AUC**:  
   The optimal threshold (0.4) maximizes **Youden's J statistic (TPR - FPR)**, providing the best balance. The ideal AUC of 1 is reached when the ROC curve is in the top-left corner, indicating perfect classification.

In summary, the optimized threshold enhances the model by improving recall and balancing false positives/negatives while maintaining overall performance.

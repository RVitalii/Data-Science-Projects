# SPAM PREDICTION

## Goal
The objective is to develop and train a binomial logistic regression model for the purpose of identifying spam.

## Model

The model belongs to the class of linear models, similar to linear regression. It is widely used for classification problems. 

Logistic regression comprises two building blocks: 
1. The simple logistic function:
\begin{equation}
p(x) = \frac{e^x}{1+e^x}.
\end{equation}
2. The linear regression (in matrix form):
\begin{equation}
Y = X\beta + \epsilon.
\end{equation}

The composition of 1-2 yields:
\begin{equation}
p(X) = \frac{e^{X\beta + \epsilon}}{1+e^{X\beta + \epsilon}}.
\end{equation}

The above result can be interpreted as a log of odds, since the model can be rewritten as: 
\begin{equation}
log\left(\frac{p(X)}{1 - p(X)}\right) = X\beta + \epsilon.
\end{equation}
Note, that the left hand side of the above equation represents the logit function -- the inverse of the logistic function. This is a quantile function of the logistic distribution. Therefore the output of the model is a conditional probability of the binary dependent variable given some random vector of independent variables, that can be either descrete or continuous.

## Binomial Logistic Regression Assumptions

1. Independent errors.
2. Errors are homoscedastic.
3. Binary dependent variables.
4. There is no multicollinearity between independent variables.
5. The dependet variables are exogenous, meaning they are not influenced by the dependent variables.
5. It is possible to establish linear relationship between independent variables and log odds.
6. There is no outliers in data.
